```python
from utils import win_loss_utils
import pandas as pd

win_loss_utils.save_data()
```

```python
all_data, home_data, away_data = win_loss_utils.load_data()
all_data["Start Date"] = pd.to_datetime(all_data["Start Date"], format="%d %b %Y")
home_data["Start Date"] = pd.to_datetime(home_data["Start Date"], format="%d %b %Y")
away_data["Start Date"] = pd.to_datetime(away_data["Start Date"], format="%d %b %Y")
```

```python
all_rolling_data = win_loss_utils.get_rolling_win_percentage(years=5, data=all_data)
home_rolling_data = win_loss_utils.get_rolling_win_percentage(years=5, data=home_data)
away_rolling_data = win_loss_utils.get_rolling_win_percentage(years=5, data=away_data)

all_rolling_data = all_rolling_data.reset_index()
home_rolling_data = home_rolling_data.reset_index()
away_rolling_data = away_rolling_data.reset_index()
```

```python
import plotly.graph_objects as go

def plot_stacked_area_chart(rolling_data):
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=rolling_data["Time"], y=rolling_data["Win"],
        hoverinfo='x+y',
        mode='lines',
        line=dict(width=0.5, color='rgb(131, 90, 241)'),
        stackgroup='one',
        name="Win"
    ))
    fig.add_trace(go.Scatter(
        x=rolling_data["Time"], y=rolling_data["Loss"],
        hoverinfo='x+y',
        mode='lines',
        line=dict(width=0.5, color='rgb(111, 231, 219)'),
        stackgroup='one',
        name="Loss"
    ))
    fig.add_trace(go.Scatter(
        x=rolling_data["Time"], y=rolling_data["Draw"],
        hoverinfo='x+y',
        mode='lines',
        line=dict(width=0.5, color='rgb(184, 247, 212)'),
        stackgroup='one',
        name="Draw"
    ))
 
    fig.update_layout(yaxis_range=(0, 1))
    fig.show()
```

## 5 Year Rolling Win/Draw/Loss Percentage (Stacked Area Charts)

```python
plot_stacked_area_chart(all_rolling_data)
plot_stacked_area_chart(home_rolling_data)
plot_stacked_area_chart(away_rolling_data)
```
